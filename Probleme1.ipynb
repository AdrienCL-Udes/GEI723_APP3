
## Équation 1
### Méthode d'euler
$$
\frac{dI_{i}}{dt} = -\frac{I_{i}(t)}{\tau_{i}} + \sum_{j=1}^{N} w_{ij}S_{j}(t) \rightarrow \frac{dI_{i}}{dt} = \frac{I_{i}(t + \Delta t) - I_{i}(t)}{\Delta t}
\\
\text{ }
\\
\frac{I_{i}(t + \Delta t) - I_{i}(t)}{\Delta t} = -\frac{I_{i}(t)}{\tau_{i}} + \sum_{j=1}^{N} w_{ij}S_{j}(t)
\\
\text{ }
\\
I_{i}(t + \Delta t) - I_{i}(t) = \Delta t\left(-\frac{I_{i}(t)}{\tau_{i}} + \sum_{j=1}^{N} w_{ij}S_{j}(t)\right)
\\
\text{ }
\\
I_{i}(t + \Delta t) - I_{i}(t) = -I_{i}(t)\frac{\Delta t}{\tau_{i}} + \Delta t\sum_{j=1}^{N} w_{ij}S_{j}(t)
\\
\text{ }
\\
I_{i}(t + \Delta t) = I_{i}(t) - I_{i}(t)\frac{\Delta t}{\tau_{i}} + \Delta t\sum_{j=1}^{N} w_{ij}S_{j}(t)
\\
\text{ }
\\
I_{i}(t + \Delta t) = I_{i}(t) \left(1 - \frac{\Delta t}{\tau_{i}}\right) + \Delta t\sum_{j=1}^{N} w_{ij}S_{j}(t)
$$
### Version numérique

```python
def sum(w, S, N):
    total = 0
    for j in range(N):
        total += w[j] * S[j]

    return total 
    
I[i][n + 1] = I[i][n] * (1 + deltaT) + deltaT * sum(w[i], S, N)
```

$$

\\
\text{ }
\\
\frac{U_{i}(t + \Delta t) - U_{i}(t)}{\Delta t} = \frac{-U_{i}(t) + I_{i}(t)}{\tau_{U}}
\\
\text{ }
\\
U_{i}(t + \Delta t) - U_{i}(t) = \frac{\Delta t}{\tau_{U}} \left(-U_{i}(t) + I_{i}(t)\right)
\\
\text{ }
\\
U_{i}(t + \Delta t) - U_{i}(t) = -U_{i}(t)\frac{\Delta t}{\tau_{U}} + I_{i}(t)\frac{\Delta t}{\tau_{U}} 
\\
\text{ }
\\
U_{i}(t + \Delta t) = U_{i}(t) - U_{i}(t)\frac{\Delta t}{\tau_{U}}  + I_{i}(t)\frac{\Delta t}{\tau_{U}}
\\
\text{ }
\\
U_{i}(t + \Delta t) = U_{i}(t)\left(1 - \frac{\Delta t}{\tau_{U}}\right) + I_{i}(t)\frac{\Delta t}{\tau_{U}}
$$
